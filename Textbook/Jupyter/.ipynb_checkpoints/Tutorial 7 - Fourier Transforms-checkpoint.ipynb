```python
fig = plt.figure(figsize=(8, 4))

def dispersion(k):
    return np.sqrt(k)

def transform(k):
    a = 0
    n = 20
    rs = np.random.RandomState(1)
    for i in range(n):
        a0 = rs.random_sample()
        k0 = 20 * rs.random_sample() + 1j * rs.random_sample()
        a += a0 * np.exp(-(k - k0)**2)
    return a

def wavepacket(t, N = 1000, x1 = -10, x2 = 10):
    
    
    kmin = -100
    kmax = 100
    dk = (kmax - kmin) / N
    k = np.arange(kmin, kmax, dk)
    a = transform(k)
    w = dispersion(np.abs(k))

    a = a * np.exp(-1j*w*t)
    
    eta = np.fft.ifft(np.fft.ifftshift(a))
    eta = np.fft.fftshift(eta)
    
    return np.real(eta)

    


eta = wavepacket(100)

ax = fig.add_subplot(1, 1, 1)

ax.plot(eta, color="black")
#ax.set_xlim(1200, 3800)
#ax.set_ylim(-0.02, 0.02)
#ax.axis('off')


```
