```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

def JT(z, c = 0.7):
    xi = z + c**2 / z
    return xi

def JT_inverse(Z, c = 0.7):
    return 0.5 * Z + np.sqrt(0.25 * Z**2 - c**2)

def circle(R = 1.0, centre = 0):
    angles = np.linspace(0, 2*np.pi, 100)
    circ = centre + R * np.exp(1j*angles)
    return circ

def w(z, R = 1.0, alpha = 0, Gamma = 0.0, lam = 0.0):
    return (z + lam)*np.exp(-1j*alpha) + (R+ lam)**2/(z+lam) * np.exp(1j * alpha) - 1j * Gamma / 2.0 / np.pi * np.log(z + lam)

a = 1.15
c = 1
lam = -0.15

radii = np.linspace(a, 5.0, 145)
angles = np.linspace(0, 2*np.pi, 145)
radii, angles = np.meshgrid(radii, angles)

fig = plt.figure(figsize=(8, 8))

ax = fig.add_subplot(1, 2, 1)
#ax.axis('off')
ax.set_aspect('equal')
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)

z =  radii * np.exp(1j * angles)
zjt = JT(z, c = c)
ax.scatter(z.real, z.imag, s=0.1)

ax = fig.add_subplot(1, 2, 2)
#ax.axis('off')
ax.set_aspect('equal')
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)

z =  lam + radii * np.exp(1j * angles)
zjt = JT(z, c = c)
ax.scatter(zjt.real, zjt.imag, s=0.1)


plt.show()
```
