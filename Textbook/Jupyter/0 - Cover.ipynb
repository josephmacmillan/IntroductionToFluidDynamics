```python
%matplotlib inline

import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np

rc('text.latex',preamble='\\usepackage{libertine}\n\\usepackage[libertine]{newtxmath}')
rc('font',**{'family':'serif','serif':['Linux Libertine O']}, size=18)
rc('text', usetex=True)
```

```python
def draw_wave(ax):

    x = np.linspace(0, 15.25, 1000)
    y = np.sin(x**1.5)

    def rotate(x, y, angle):
        xp = x * np.cos(angle) - y * np.sin(angle)
        yp = x * np.sin(angle) + y * np.cos(angle)

        return xp, yp

    ax.plot(x, y, lw=2, color='#ffb703ff')
    for a in np.arange(-0.03, 0.03, 0.0025):
        xp, yp = rotate(x, y, a)
        ax.plot(xp, yp, lw=1, color='#ffb703ff')

    
fig = plt.figure(figsize=(6, 9))
ax = fig.add_axes([0, 0, 1, 1])



x = np.linspace(0, 15.25, 1000)
y = np.sin(x**1.5)
ax.fill_between(x, y, y2 = -40, color='#219ebcff')
ax.fill_between(x, y, y2 = 30, color='#8ecae6ff')

draw_wave(ax)

ax.set_ylim(-35, 25)
ax.set_xlim(0, 15.25)
ax.axis('off')

#ax.text(1, 5, "Introduction to Fluid Dynamics", fontsize=30)
#ax.text(4.5, -5, "Joseph D. MacMillan", fontsize=20)

plt.savefig('fig_cover1.pdf')
```

```python
def draw_wave(ax, color='#002f8aff'):

    x = np.linspace(0, 15.25, 1000)
    y = np.sin(x**1.5)

    def rotate(x, y, angle):
        xp = x * np.cos(angle) - y * np.sin(angle)
        yp = x * np.sin(angle) + y * np.cos(angle)

        return xp, yp

    ax.plot(x, y, lw=2, color=color)
    for a in np.arange(-0.03, 0.03, 0.0025):
        xp, yp = rotate(x, y, a)
        ax.plot(xp, yp, lw=1, color=color)

    
fig = plt.figure(figsize=(6, 0.5))
ax = fig.add_axes([0, 0, 1, 1])

draw_wave(ax, color="palegoldenrod")
ax.axis('off')

plt.savefig('fig_cover_wave.pdf')
```
