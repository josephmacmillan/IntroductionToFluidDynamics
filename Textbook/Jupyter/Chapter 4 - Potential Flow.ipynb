## Figure 4.2 - Basic Potential Flows

Plots of the four basic potential flows:
* Uniform flow, $w(z) = Uze^{-i\alpha}$.
* Stagnation point, $w(z) = \frac{1}{2} \alpha z^2$.
* Line vortex, $w(z) = -\frac{i\Gamma}{2\pi} \ln z$.
* Line source, $w(z) = \frac{Q}{2\pi} \ln z$.

```python
def uniform(z, U = 1.0, alpha = np.pi/6):
    
    return U * z * np.exp(-1j * alpha)

def stagnation(z, alpha = 1.0):

    return 0.5 * alpha * z**2

def vortex(z, Gamma = 1.0, d = 0.0):

    return -1j * Gamma / 2 / np.pi * np.log(z - d)

def source(z, Q = 1.0):
    
    return Q / 2 / np.pi * np.log(z)

xx = np.linspace(-5.0, 5.0, 1000)
yy = np.linspace(-5.0, 5.0, 1000)
xx, yy = np.meshgrid(xx, yy)

zz = xx + 1j * yy

fig = plt.figure(figsize=(6, 6))

ax = fig.add_subplot(2, 2, 1)
ax.axis('off')
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
ax.contour(xx, yy, uniform(zz).imag, colors='black', levels=20)

ax = fig.add_subplot(2, 2, 2)
ax.axis('off')
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
ax.contour(xx, yy, stagnation(zz).imag, colors='black', levels=20)

ax = fig.add_subplot(2, 2, 3)
ax.axis('off')
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
ax.contour(xx, yy, vortex(zz).imag, colors='black', levels=20)

ax = fig.add_subplot(2, 2, 4)
ax.axis('off')
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
ax.contour(xx, yy, source(zz).imag, colors='black', levels=20)

plt.savefig("fig_basic_flows.pdf", bbox_inches='tight')
```

## Figure 4.6 - Line vortex beside a wall

The complex potential for a line vortex a distance $d$ from a wall at $x=0$ is
$$
w(z) = -\frac{i\Gamma}{2\pi} \ln \left( \frac{z - d}{z+d} \right).
$$
Here are the streamlines.

```python
def vortex_pair(z, Gamma1 = 1.0, Gamma2 = -1.0, d1 = 1.0, d2 = -1.0):
    
    return vortex(z, Gamma1, d1) + vortex(z, Gamma2, d2)


fig = plt.figure(figsize=(8, 4))

ax = fig.add_subplot(1, 2, 1)
ax.axis('off')
ax.set_aspect('equal')
ax.contour(xx, yy, vortex_pair(zz).imag, colors='black', levels=51, zorder=1)
ax.set_xlim(-4.1, 4.1)
ax.set_ylim(-3, 3)

ax = fig.add_subplot(1, 2, 2)
ax.axis('off')
ax.set_aspect('equal')
ax.imshow([[0, 1], [0, 1]], extent=[-4.11, 0, -3, 3], interpolation = 'bicubic', cmap='binary', alpha=1, zorder=2, vmin = 0.1, vmax = 2)
ax.contour(xx, yy, vortex_pair(zz).imag, colors='black', levels=51, zorder=1)
ax.set_xlim(-4.1, 4.1)
ax.set_ylim(-3, 3)

plt.savefig("fig_vortex_wall.pdf", bbox_inches='tight')
```

## Figure 4.8 - Vortex in a Bucket - Streamlines

The streamlines of a line vortex in a cylindrical bucket.  The complex potential is given by
$$
w(z) = -\frac{i\Gamma}{2\pi} \ln (z - c) + \frac{i\Gamma}{2\pi} \ln (z - a^2/c),
$$
where $a$ is the radius of the cylinder and $c$ is the distance of the vortex from the centre.

```python
fig = plt.figure(figsize=(6, 6))

ax = fig.add_subplot(1, 1, 1)
ax.axis('off')
ax.set_aspect('equal')
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)

a = 1.0
c1 = 0.4
c2 = a**2/c1

rect = matplotlib.patches.Rectangle((-1.5, -1.5), 3, 3, facecolor='lightgrey', zorder=0)
circ = plt.Circle((0,0),radius=a, fc='white', linewidth=2.5, fill=True, alpha=1, zorder=1, color="black")
ax.add_patch(rect)
ax.add_patch(circ)

ax.contour(xx, yy, vortex_pair(zz, d1 = c1, d2 = c2).imag, colors='black', levels=np.linspace(0.155, 1, 20), zorder=2)

plt.savefig("fig_vortex_cylinder.pdf", bbox_inches='tight')
```

## Figure 4.9

Using the Circle Theorem to find flow that is uniform at inifinty with a circular streamline centred at the origin.  The complex potential is
$$
w(z) = U \left( z + \frac{a^2}{z} \right).
$$


```python
def flow_around_circle(z, a = 1.0, U = 1.0, alpha = 0, Gamma = 0):
    return U * (z*np.exp(-1j*alpha) + a**2/z * np.exp(1j*alpha)) - 1j * Gamma / 2.0 / np.pi * np.log(z)

fig = plt.figure(figsize=(6, 4))

ax = fig.add_subplot(1, 1, 1)
ax.axis('off')
ax.set_aspect('equal')
ax.set_xlim(-3, 3)
ax.set_ylim(-2.1, 2.1)

ax.contour(xx, yy, flow_around_circle(zz).imag, colors='black', levels=np.linspace(-10, 10, 41))

plt.savefig("fig_uniform_circle.pdf", bbox_inches='tight')
```

## Figure 4.11 - A conformal map


```python
def conformal_map(z):
    return z**2 / 5 + 2

fig = plt.figure(figsize=(6, 4))

ax = fig.add_subplot(1, 2, 1)
ax.axis('off')
ax.set_aspect('equal')
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)

ax.contour(xx, yy, uniform(zz, alpha=0).imag, colors='blue', levels=np.linspace(-6, 6, 21))
ax.contour(xx, yy, uniform(zz, alpha=0).real, colors='red', levels=np.linspace(-6, 6, 21))

xt = conformal_map(zz).real
yt = conformal_map(zz).imag

ax = fig.add_subplot(1, 2, 2)
ax.axis('off')
ax.set_aspect('equal')
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.contour(xt, yt, uniform(zz, alpha=0).imag, colors='blue', levels=np.linspace(-6, 6, 21))
ax.contour(xt, yt, uniform(zz, alpha=0).real, colors='red', levels=np.linspace(-6, 6, 21))

plt.savefig("fig_conformal_map.pdf", bbox_inches='tight')
```

## Figure 4.12

The Joukowski transformation,
$$
Z = f(z) = z + \frac{c^2}{z},
$$
applied to a circle of radius $a$,
$$
z = a e^{i\phi}.
$$

```python
a = 1.0
c = 0.7
w = 2*(a + c**2 / a)
h = 2*(a - c**2 / a)

fig = plt.figure(figsize=(6, 4))

ax = fig.add_subplot(1, 1, 1)
ax.axis('off')
ax.set_aspect('equal')
ax.set_xlim(-2, 6)
ax.set_ylim(-2, 2)

ax.add_patch(plt.Circle((0,0),radius=a, fc='lightgrey', fill=True, alpha=1, zorder=2, color="black"))
ax.add_patch(matplotlib.patches.Ellipse((4,0), w, h,fc='lightgrey', fill=True, alpha=1, zorder=2, color="black"))
ax.add_patch(matplotlib.patches.FancyArrowPatch((1.2, 0), (2.3, 0), arrowstyle="Simple, tail_width=0.5, head_width=3, head_length=3", mutation_scale=5, fc='black'))

plt.savefig("fig_jt_ellipse.pdf", bbox_inches='tight')
```

## Figure 4.13

Streamlines around an elliptical cylinder, with complex potential
$$
W(Z) = Ue^{-i\alpha} \left( \frac{1}{2} Z + \sqrt{\frac{1}{4} Z^2 - c^2} \right) + Ue^{i\alpha} \frac{a^2}{c^2} \left( \frac{1}{2} Z - \sqrt{\frac{1}{4} Z^2 - c^2} \right) - \\ \frac{i\Gamma}{2\pi} \ln \left( \frac{1}{2} Z + \sqrt{\frac{1}{4} Z^2 - c^2} \right)
$$

```python
def joukowski(z, c = 0.7):
    return z + c**2 / z 

def circle(z, a = 1.0, centre = 0.0):
    angles = np.linspace(0, 2*np.pi, 100)
    circ = centre + a * np.exp(1j*angles)
    return circ

radii = np.linspace(1.01, 5.0, 145)
angles = np.linspace(0, 2*np.pi, 145)
radii, angles = np.meshgrid(radii, angles)

zr =  radii * np.exp(1j * angles)

fig = plt.figure(figsize=(6, 4))

ax = fig.add_subplot(1, 1, 1)
ax.axis('off')
ax.set_aspect('equal')
ax.set_xlim(-3, 3)
ax.set_ylim(-2, 2)

xt = joukowski(zr).real
yt = joukowski(zr).imag

ax.contour(xt, yt, flow_around_circle(zr, alpha=np.pi/6, Gamma = 0).imag, colors='black', levels=np.linspace(-5, 5, 30))

circ = joukowski(circle(zr))
ax.fill(circ.real, circ.imag, fc="lightgrey", ec="black")

plt.savefig("fig_ellipse.pdf", bbox_inches='tight')
```

## Figure 4.16 - Streamlines and pressure around an aerofoil

```python
def flow_around_shifted_circle(z, a = 1.0, U = 1.0, alpha = 0, Gamma = 0, lam = 0.0):
    return U * ((z+lam)*np.exp(-1j*alpha) + (a+lam)**2/(z+lam) * np.exp(1j*alpha)) - 1j * Gamma / 2.0 / np.pi * np.log(z+lam)

def vel_around_shifted_circle(z, a = 1.0, U = 1.0, alpha = 0, Gamma = 0, lam = 0.0):
    return U * ( np.exp(-1j * alpha) - ((a+lam)/(z+lam))**2 * np.exp(1j*alpha) ) - 1j*Gamma / (2.0 * np.pi * (z+lam))

def joulowski_deriv(z, c = 0.7):
    return 1 - c**2/z**2

fig = plt.figure(figsize=(6, 4))

ax = fig.add_subplot(1, 1, 1)
ax.axis('off')
ax.set_aspect('equal')
ax.set_xlim(-4, 4)
ax.set_ylim(-2, 2)

lam = 0.2
Gamma = -4.0*np.pi*(a+lam)*np.sin(np.pi/6)
alpha = np.pi/6

radii = np.linspace(a+lam, 5.0, 145)
angles = np.linspace(0, 2*np.pi, 145)
radii, angles = np.meshgrid(radii, angles)
zr = -lam + radii * np.exp(1j * angles)

psi = flow_around_shifted_circle(zr, a, lam=lam, alpha = alpha, Gamma = Gamma).imag
zjt = joukowski(zr, c = a)
ax.contour(zjt.real, zjt.imag, psi, colors='white', levels=50, linewidths=1)

dw = vel_around_shifted_circle(zr, a, lam=lam, alpha = alpha, Gamma = Gamma) / joulowski_deriv(zr, c=0.99)
u = dw.real
v = -dw.imag
uu = u**2 + v**2
p = 0.5 * (1 - uu)

ax.contourf(zjt.real, zjt.imag, p, cmap = 'coolwarm', levels=1000, vmin = -4)
aerofoil = joukowski(circ, c = a)
ax.fill(aerofoil.real, aerofoil.imag, fc="lightgrey", ec="black")

plt.savefig("fig_aerofoil_stream.pdf", bbox_inches='tight')
```

## Figure 4.22 - A Vortex Pair

A vortex pair moving downstream, in the reference frame of the pair.  The complex potential is
$$
w(z) = -\frac{i\Gamma z}{4\pi d} - \frac{i\Gamma}{2\pi} \ln (z-d) + \frac{i\Gamma}{2\pi} \ln (z+d).
$$

```python
Gamma = 1.0
d = 1.0
w = -1j*Gamma * zz / (4.0 * np.pi * d) -1j * Gamma / (2.0 * np.pi) * np.log(zz - d) + 1j * Gamma / (2.0 * np.pi) * np.log(zz + d)

fig = plt.figure(figsize=(6, 4))

ax = fig.add_subplot(1, 1, 1)
ax.axis('off')
ax.set_aspect('equal')
ax.set_xlim(-4.1, 4.1)
ax.set_ylim(-3, 3)

ax.contour(xx, yy, w.imag, levels=41, colors="black")

plt.savefig("fig_vortex_pair3.pdf", bbox_inches='tight')
```

## Figure 4.25 - Attached Vortexes

The complex potential is given by
$$
w(z) = \frac{1}{2} \alpha z^2 + \frac{i\Gamma}{2\pi} \ln (z - z_0) -  \frac{i\Gamma}{2\pi} \ln (z - z_0^*) -  \frac{i\Gamma}{2\pi} \ln (z + z_0^*) +  \frac{i\Gamma}{2\pi} \ln (z + z_0)
$$
where
$$ 
z_0 = d + id
$$
and for fixed vortexes
$$
d = \sqrt{\frac{\Gamma}{8 \pi \alpha}}.
$$


```python
Gamma = 1.0
alpha = 1.0
d =np.sqrt(Gamma / 8 / np.pi / alpha)
z0 = d + 1j*d
z0star = d - 1j*d

xx = np.linspace(0.0, 2.0, 1000)
yy = np.linspace(-2.0, 2.0, 1000)
xx, yy = np.meshgrid(xx, yy)

zz = xx + 1j * yy

w = alpha * zz**2 / 2 + 1j * Gamma / 2 / np.pi * np.log(zz - z0) - 1j * Gamma / 2 / np.pi * np.log(zz - z0star) - 1j * Gamma / 2 / np.pi * np.log(zz + z0star) + 1j * Gamma / 2 / np.pi * np.log(zz + z0)
dw = alpha * zz + 1j * Gamma / 2 / np.pi / (zz - z0) - 1j * Gamma / 2 / np.pi / (zz - z0star) - 1j * Gamma / 2 / np.pi / (zz + z0star) + 1j * Gamma / 2 / np.pi / (zz + z0)

fig = plt.figure(figsize=(6, 4))

ax = fig.add_subplot(1, 1, 1)
ax.axis('off')
ax.set_aspect('equal')
ax.set_xlim(-0.4, 1.2)
ax.set_ylim(-0.8, 0.8)

ax.contour(xx, yy, w.imag, levels=np.linspace(-2, 2, 50), colors="white")

u = dw.real
v = -dw.imag
uu = u**2 + v**2
p = 0.5 * (1 - uu)
p[p<-1] = -1

cont = ax.contourf(xx, yy, p, cmap = 'coolwarm', levels=np.linspace(-1, 1, 100))
#print(cont.levels)

ax.imshow([[0, 1], [0, 1]], extent=[-0.4, 0, -0.8, 0.8], interpolation = 'bicubic', cmap='binary', alpha=1, zorder=2, vmin = 0.1, vmax = 2)

plt.savefig("fig_attached_vortex2.pdf", bbox_inches='tight')
```

## Figure 4.28 - Vortex Street

Streamlines for the vortex street model, with compelx potential
$$
w(z) = -\frac{i\Gamma}{2\pi} \ln \left[ \sin \left( \frac{\pi z}{a} \right) \right].
$$

```python
Gamma = 1.0
a = 1.0
b = 0.3
U = Gamma / 2 / a * np.tanh(np.pi * b / a)

xx = np.linspace(-5.0, 5.0, 1000)
yy = np.linspace(-5.0, 5.0, 1000)
xx, yy = np.meshgrid(xx, yy)

zz = xx + 1j * yy

w = -1j * Gamma / 2 / a * np.log( np.sin(np.pi * zz / a)) +1j * Gamma / 2 / a * np.log( np.sin(np.pi * (zz - a/2 - 1j*b) / a )) + U * zz
dw = -1j * Gamma / 2 / a / np.tan(np.pi * zz / a) +1j * Gamma / 2 / a / np.tan(np.pi * (zz - a/2 - 1j*b) / a) + U

fig = plt.figure(figsize=(8, 4))

ax = fig.add_subplot(1, 1, 1)
ax.axis('off')
ax.set_aspect('equal')
ax.set_xlim(-2, 2)
ax.set_ylim(-1, 1.3)

print(w.imag.max(), w.imag.min())

ax.contour(xx, yy, w.imag, levels=np.linspace(-3, 3, 51), colors="white", linewidths=1)

u = dw.real
v = -dw.imag
uu = u**2 + v**2
p = 0.5 * (1 - uu)
print(p.max(), p.min())
p[p<-0.5] = -0.5

ax.contourf(xx, yy, p, cmap = 'coolwarm', levels=100)

plt.savefig("fig_vortex_street2.pdf", bbox_inches='tight')
```
